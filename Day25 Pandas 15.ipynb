```python
# # Self defined function
# 1) Calculate total marks for each student
# 2) Calculate  percentage for each student
# 3) Give Grade according to marks 
# Above 90% Grade "O"
# Above 80% Grade "A+"
# Above 70% Grade "A"
# Above 60% Grade "B+"
# Above 50% Grade "B"
# Above 40% Grade "C+"
# Above 35% Grade "C"
# Otherwise Grade "F"
```

```python
def Calculate_Result(data):
    data["Total Marks"]=data["English"]+data["Physics"]+data["Chemistry"]+data["Mathematics"]
    data["Percentage"]=(data["Total Marks"]/400)*100
    if data["Percentage"]>90:
        data["Grade"]="O"
    elif data["Percentage"]>80:
        data["Grade"]="A+"
    elif data["Percentage"]>70:
        data["Grade"]="A"        
    elif data["Percentage"]>60:
        data["Grade"]="B+"        
    elif data["Percentage"]>50:
        data["Grade"]="B"         
    elif data["Percentage"]>40:
        data["Grade"]="C"         
    elif data["Percentage"]>35:
        data["Grade"]="C+"         
    else:
        data["Grade"]="F"   
        
    return data
```
