```python
import pandas as pd
```

```python
# Read excel file in pandas
# Syntax
# pd.read_excel(r"file_path/file_name.xlsx") ==> Excel
# pd.read_csv(r"file_path/file_name.csv")  ==> Csv
```

```python
# path
# extensions ==> .xlsx = excel file  
                 # .csv= csv file
# single / double slash / raw string

# sheet_name : Specifies which sheet in the Excel file to read. By default read first sheet
# header (None/other number) : Define which row to use as the header (column names). # header=None==> all data consider as record
# usecols : Define which columns to read from the Excel sheet.
```

```python
# skiprows : Skips a specified number of rows at the start ==> int ==> # starting recods removed
#                                                          ==> []  ==> Defined rows removed
```
