```python
# Filter data : Students secured greater than 80 marks in Physics
data2[data2["Physics"]>80]
```

```python
# Filter data : Students name is "Samarth"
data2[data2["First Name"]=="Samarth"]
```

```python
# Filter records whoes First name is Arjun and Tarun using operator
data2[(data2["First Name"]=="Arjun") | (data2["First Name"]=="Tarun")] 
```

```python
data2[(data2["First Name"]=="Arjun") & (data2["First Name"]=="Tarun")]  # We can not use and operator to filter out multiple values from single column
```

```python
# Filter records whoes First name is Arjun and Tarun using isin
data2[data2["First Name"].isin(["Arjun","Tarun"])]
```

```python
data2[data2["First Name"].isin(["Arjun","Tarun","Aarav"])]
```

```python
# First name except Arjun and Tarun
data2[~(data2["First Name"].isin(["Arjun","Tarun"]))]
```
